# Pyplot
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

# Matploting line:
it is used to ploting the value in graph in a straight line .

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

# Matplotlib Labels and Title

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,6])
y = np.array([0,250])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

#Matplotlib Scatter

With Pyplot, you can use the scatter() function to draw a scatter plot.

function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

# Matplotlib Bars

With Pyplot, you can use the bar() function to draw bar graphs:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

# Matplotlib Histograms

It is a graph showing the number of observations within each given interval.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.hist(x)
plt.show()
```

# Matplotlib.pie charts

With Pyplot, you can use the pie() function to draw pie charts:

```python

import matplotlib.pyplot as plt
import numpy as np

sized = np.array([35, 25, 25, 15])
labels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(sized, labels=labels,autopct="%1.1f%%")
plt.title('fruits')
plt.show()
```

# Matplotlib Areaplot

to identifly differeance b/w the tow data plots

```python
import matplotlib.pyplot as plt
x =[1,2,3,4,5]
y1,y2 = [10,80,30,80,50],[5,20,25,35,50]
plt.fill_between(x,y1,y2,color = 'blue',alpha = 0.6,label ='area b/w 1-2')
plt.plot(x,y1,label = '1st',marker = '*')
plt.plot(x,y2,label = '2nd',marker = 'o')
plt.legend()
plt.show()
```

# Introduction of Figures



```python
fig = plt.figure(figsize=(5,5))
fig.suptitle('figure ')
d = fig.add_subplot(111)
d.plot([1,2,3])
a = fig.add_subplot(221,facecolor = 'blue')
a.plot([1,2,3])
d.set_title('subplot')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,a= plt.subplots(2,2)

import numpy as np
x=np.arange(4,10)
a[0,0].plot(x,x*x)
a[0,0].set_title('square')
a[0,1].plot(x,np.sqrt(x))
a[0,1].set_title('square root')
a[1,0].plot(x,np.log(x))
a[1,0].set_title('log')
a[1,1].plot(x,np.exp(x))
a[1,1].set_title('exp')
plt.show()
```
